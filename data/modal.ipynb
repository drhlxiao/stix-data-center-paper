```python
import pandas as pd
df=pd.read_csv('cal.csv',delimiter='\t')
data=df.to_numpy()
#print(data)
x,y=data.T
from matplotlib import pyplot as plt
plt.figure(figsize=(4,3))
plt.plot(x,y/max(y))
plt.xlabel('Energy (keV)')
plt.ylabel('Arb. Units')
plt.tight_layout()
plt.savefig('cal-model.pdf')
```
